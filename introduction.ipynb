```python
from improc import PGM
import numpy as np
import matplotlib.pyplot as plt
```

## Functions to apply each pixel

```python
def img_negative(x: int) -> int:
    return 255 - x

def img_log(x: int) -> int:
    c = 1
    nx: float = x / 255 # normalizing x
    result: float = c * np.log2(1 + nx) * 255 
    return int(result)

def img_inverse_log(x: int) -> int:
    A: float = 1 / (np.e - 1)
    nx: float = x / 255 # normalized x
    result: float = A * (np.exp(nx) - 1) * 255

    return int(result)

x_axis = np.linspace(0, 255, 50)

y_id = x_axis # f(x) = x
y_lin = np.array([img_negative(i) for i in x_axis]) 
y_log = np.array([img_log(i) for i in x_axis])
y_invlog = np.array([img_inverse_log(i) for i in x_axis])

plt.figure(figsize=(10, 10))
plt.scatter(x_axis, y_id, label='Identity')
plt.scatter(x_axis, y_lin, label='Negative')
plt.scatter(x_axis, y_log, label='Log, c=1')
plt.scatter(x_axis, y_invlog, label='Inv Log')
plt.xlabel('input')
plt.ylabel('output')
plt.grid(True)
plt.legend()
plt.show()
```

# Figure 3.4

### Image Loading

```python
img_path: str = "./images/"

pgm3_4 = PGM.read("".join([img_path, "fig-3_4.pgm"]))

pgm3_4_original = pgm3_4.to_ndarray()
pgm3_4_lin = pgm3_4.map(img_negative).to_ndarray()
pgm3_4_log = pgm3_4.map(img_log).to_ndarray()
pgm3_4_ilog = pgm3_4.map(img_inverse_log).to_ndarray()
```

### Plotting

```python
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.imshow(pgm3_4_original, cmap='gray', vmin=0, vmax=255)
plt.title("Original")

plt.subplot(2, 2, 2)
plt.imshow(pgm3_4_lin, cmap='gray', vmin=0, vmax=255)
plt.title("Negative")

plt.subplot(2, 2, 3)
plt.imshow(pgm3_4_log, cmap='gray', vmin=0, vmax=255)
plt.title("Log")

plt.subplot(2, 2, 4)
plt.imshow(pgm3_4_ilog, cmap='gray', vmin=0, vmax=255)
plt.title("Inverse Log")

plt.show()
```
