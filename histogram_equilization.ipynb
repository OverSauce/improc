```python
from improc import PGM
import matplotlib.pyplot as plt
import numpy as np
```

# Figure 3.20

Loading Files

```python
img_path: str = "./images/"

pgm3_20_a = PGM.read("".join([img_path, "fig-3_20-a.pgm"]))
pgm3_20_b = PGM.read("".join([img_path, "fig-3_20-b.pgm"]))
pgm3_20_c = PGM.read("".join([img_path, "fig-3_20-c.pgm"]))
pgm3_20_d = PGM.read("".join([img_path, "fig-3_20-d.pgm"]))
```

#### Loading original images

```python
pgm3_20_a_original = pgm3_20_a.to_ndarray()
pgm3_20_b_original = pgm3_20_b.to_ndarray()
pgm3_20_c_original = pgm3_20_c.to_ndarray()
pgm3_20_d_original = pgm3_20_d.to_ndarray()
```

#### Histogram equilized images

```python
pgm3_20_a_equalized = pgm3_20_a.hist_eq().to_ndarray()
pgm3_20_b_equalized = pgm3_20_b.hist_eq().to_ndarray()
pgm3_20_c_equalized = pgm3_20_c.hist_eq().to_ndarray()
pgm3_20_d_equalized = pgm3_20_d.hist_eq().to_ndarray()
```

#### Normalized histograms

```python
pgm3_20_a_histogram = pgm3_20_a.histogram(normalize=True)
pgm3_20_b_histogram = pgm3_20_b.histogram(normalize=True)
pgm3_20_c_histogram = pgm3_20_c.histogram(normalize=True)
pgm3_20_d_histogram = pgm3_20_d.histogram(normalize=True)
```

### Plots

```python
plt.figure(figsize=(15, 15))

# Figure 3.20-a

plt.subplot(4, 3, 1)
plt.imshow(pgm3_20_a_original, cmap="gray", vmin=0, vmax=255)
plt.title("Figure 3.20-a - Original")

plt.subplot(4, 3, 2)
plt.imshow(pgm3_20_a_equalized, cmap="gray", vmin=0, vmax=255)
plt.title("Figure 3.20-a - Equalized")

plt.subplot(4, 3, 3)
plt.plot(pgm3_20_a_histogram)
plt.title("Figure 3.20-a - Histogram")

# Figure 3.20-b

plt.subplot(4, 3, 4)
plt.imshow(pgm3_20_b_original, cmap="gray", vmin=0, vmax=255)
plt.title("Figure 3.20-b - Original")

plt.subplot(4, 3, 5)
plt.imshow(pgm3_20_b_equalized, cmap="gray", vmin=0, vmax=255)
plt.title("Figure 3.20-b - Equalized")

plt.subplot(4, 3, 6)
plt.plot(pgm3_20_b_histogram)
plt.title("Figure 3.20-b - Histogram")

# Figure 3.20-c

plt.subplot(4, 3, 7)
plt.imshow(pgm3_20_c_original, cmap="gray", vmin=0, vmax=255)
plt.title("Figure 3.20-c - Original")

plt.subplot(4, 3, 8)
plt.imshow(pgm3_20_c_equalized, cmap="gray", vmin=0, vmax=255)
plt.title("Figure 3.20-c - Equalized")

plt.subplot(4, 3, 9)
plt.plot(pgm3_20_c_histogram)
plt.title("Figure 3.20-c - Histogram")

# Figure 3.20-d

plt.subplot(4, 3, 10)
plt.imshow(pgm3_20_d_original, cmap="gray", vmin=0, vmax=255)
plt.title("Figure 3.20-d - Original")

plt.subplot(4, 3, 11)
plt.imshow(pgm3_20_d_equalized, cmap="gray", vmin=0, vmax=255)
plt.title("Figure 3.20-d - Equalized")

plt.subplot(4, 3, 12)
plt.plot(pgm3_20_d_histogram)
plt.title("Figure 3.20-d - Histogram")

plt.show()
```
